```python
from llama_cpp import Llama

llm = Llama(
    model_path="/kaggle/input/report-gen/transformers/default/1/llama-3-8b-instruct.Q4_K_M.gguf",
    n_gpu_layers=-1,
    n_ctx=4096,
)
```

```python
# ----------------------------
# WARM UP MODEL
# ----------------------------
subtasks = [
   {
    "name":"Identify Stakeholder List",
    "status":"completed",
   },
   {
    "name":"Schedule Interview Sessions",
    "status":"in_progress",
   },
   {
    "name":"Document Interview Findings",
    "status":"not_started",
   },
]
project="HealthTrack Mobile App"
project_desc="A secure mobile application that allows patients to monitor vital signs, schedule appointments, and access electronic health records."
phase="Project Initiation & Requirements Gathering"
phase_desc="Define project scope, objectives, and stakeholder expectations. Conduct workshops with healthcare professionals, patients, and IT staff to capture functional and non‑functional requirements, including data privacy and regulatory compliance."
collection="Stakeholder Requirement Workshops"
collection_desc="Facilitate structured workshops with healthcare professionals, patients, and IT staff to elicit detailed functional and non‑functional requirements."
creator_name=""
employee_name=""
task_name="Conduct Stakeholder Interviews"
task_desc="rrange and conduct interviews with key stakeholders to gather detailed insights on user needs, expectations, and pain points for the HealthTrack Mobile App."
subtasks_name="Identify Stakeholder List"
subtasks_desc="Compile a comprehensive list of stakeholders including patients, healthcare providers, administrators, and regulatory bodies."


# 1. Daily Report
print("\n-------------\nDAILY REPORT\n")
daily_temperature = 0.6
daily_max_token = 250
daily_system_prompt = "You are a software engineer reporting task progress to your manager."
subtasks_info = "\n".join([f"- {s.get('name', 'N/A')} - Status: {s.get('status', 'N/A')}" for s in subtasks])
daily_user_prompt = f"""
Generate DAILY_PROGRESS_NOTE based on the following project context:\n\n
Start with Dear Mr of Ms {creator_name}, end with {employee_name}
Project: {project} ({project_desc})\n
Phase: {phase} ({phase_desc})\n
Collection: {collection} ({collection_desc})\n
Task: {task_name} ({task_desc})\n
Subtasks Status:\n{subtasks_info}"
"""
response = llm.create_chat_completion(
    messages=[
        {"role": "system", "content": daily_system_prompt},
        {"role": "user", "content": daily_user_prompt}
    ],
    max_tokens=daily_max_token,
    temperature=daily_temperature,
)


# 2. Processing Report
print("\n-------------\PROCESSING REPORT\n")
processing_temperature = 0.5
processing_max_token = 250
processing_system_prompt = "You are a software engineer reporting task progress to your manager."
processing_user_prompt = f"""
Start with Dear Mr of Ms {creator_name}, end with {employee_name}\n
Generate IN_PROGRESS_EMAIL based on the following task context:\n\n
Project: {project} ({project_desc})\n
Phase: {phase} ({phase_desc})\n
Collection: {collection} ({collection_desc})\n
Task: {task_name} ({task_desc})\n
f"Subtask: {subtasks_name} ({subtasks_desc})"
"""
response = llm.create_chat_completion(
    messages=[
        {"role": "system", "content": processing_system_prompt},
        {"role": "user", "content": processing_user_prompt}
    ],
    max_tokens=processing_max_token,
    temperature=processing_temperature,
)


# 3. Completed Report
print("\n-------------\COMPLETED REPORT\n")
completed_temperature = 0.4
completed_max_token = 250
completed_system_prompt = "You are a software engineer writing completion reports."
completed_user_prompt = f"""
Start with Dear Mr. of Ms. {creator_name}, end with {employee_name}\n
Generate COMPLETED_EMAIL based on the following task context:\n\n
Project: {project} ({project_desc})\n
Phase: {phase} ({phase_desc})\n
Collection: {collection} ({collection_desc})\n
Task: {task_name} ({task_desc})\n
f"Subtask: {subtasks_name} ({subtasks_desc})"
"""
response = llm.create_chat_completion(
    messages=[
        {"role": "system", "content": completed_system_prompt},
        {"role": "user", "content": completed_user_prompt}
    ],
    max_tokens=completed_max_token,
    temperature=completed_temperature,
)
```

```python
%%writefile server.py
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from typing import List, Optional, Dict, Any
from llama_cpp import Llama
import traceback

app = FastAPI()

# ----------------------------
# LOAD MODEL
# ----------------------------
llm = None
model_load_error = None

try:
    llm = Llama(
        model_path="/kaggle/input/report-gen/transformers/default/1/llama-3-8b-instruct.Q4_K_M.gguf",
        chat_format="llama-3",
        n_ctx=4096,
        n_gpu_layers=-1,
        verbose=False,
    )
except Exception as e:
    model_load_error = str(e)


# ----------------------------
# Pydantic MODEL for ChatCompletion payload
# ----------------------------
class ChatMessage(BaseModel):
    role: str
    content: str

class ChatCompletionRequest(BaseModel):
    model: Optional[str] = None
    messages: List[ChatMessage]
    temperature: Optional[float] = 0.2
    top_p: Optional[float] = 0.95
    max_tokens: Optional[int] = 200


# ----------------------------
# MAIN API 
# ----------------------------
@app.post("/generate-report")
def generate_report(body: ChatCompletionRequest):

    if model_load_error:
        return {
            "success": False,
            "error_type": "MODEL_LOAD_ERROR",
            "message": model_load_error
        }

    try:
        result = llm.create_chat_completion(
            model=body.model,
            messages=[m.dict() for m in body.messages],
            max_tokens=body.max_tokens,
            temperature=body.temperature,
            top_p=body.top_p,
        )

        output = result["choices"][0]["message"]["content"].strip()

        return {
            "success": True,
            "report": output,
            "raw": result
        }

    except Exception as e:
        return {
            "success": False,
            "error_type": "GENERATION_ERROR",
            "message": str(e),
            "trace": traceback.format_exc()
        }


# ----------------------------
# HEALTH CHECK
# ----------------------------
@app.get("/health")
def health():
    if model_load_error:
        return {"success": False, "error": model_load_error}
    return {"success": True, "status": "Model is ready"}
```
